```python
#function for running sql select queries
def mysql_dml_execute(query):
    mydb = sql_conn.connect(host='localhost', user='root', passwd='Root@2021', database = 'ineuron_task')
    cursor = mydb.cursor()
    cursor.execute(query)
    for i in cursor.fetchall():
        print(i)
    cursor.close()
    mydb.commit()
```
